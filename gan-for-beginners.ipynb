# Create an artificial Dataset

Instead of a collection of images, Iâ€™ll produce a description of the curves I am interested in:   
sinusoids may be mathematically described as the graph of functions **a sin(bx+c)**,  
where a, b, c are parameters which determine the height, frequency and phase of the curve respectively.   
Some examples of such curves are plotted in the following picture, produced via a Python snippet.

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import randint, uniform
 
X_MIN = -5.0    # maximum X value
X_MAX = 5.0     # minimum X value
SAMPLE_LEN = 64 # number of X values
NUMBER_OF_LINES = 4
X_COORDS = np.linspace(X_MIN , X_MAX, SAMPLE_LEN)
fig, axis = plt.subplots(1, 1)
for i in range(NUMBER_OF_LINES):
    axis.plot(X_COORDS, uniform(0.1,2.0)*np.sin(uniform(0.2,2.0)*X_COORDS + uniform(2)))
```
